```python
df['Age'].fillna(df['Age'].median(),inplace=True)

df
```

```python
df['Age']=df['Age'].astype(int)
df
```

```python
df
```

```python
df['FirstName'].fillna('Unknown', inplace=True)
df['LastName'].fillna('Unknown', inplace=True)
df= df.drop_duplicates(subset=['FirstName', 'LastName'])
df.reset_index(drop=True, inplace=True)  #resets the index to a new one and removes the old index values
df.to_csv("customers_clean.csv", index=False)
df
```

```python
df.isnull().sum()
```

```python
pip install gender_guesser
```

```python
import pandas as pd
import gender_guesser.detector as gender

detector = gender.Detector()
```

```python
df['Gender_predicted'] = df['FirstName'].apply(lambda x: detector.get_gender(x) if pd.notnull(x) else None)
#for each name,if it is not null,the detector.get_gender(x) function is used to determine the gender,if the name is missing,its return none.
```

```python
df['Gender_predicted'] = df['Gender_predicted'].replace({
    'male': 'Male',
    'mostly_male': 'Male',
    'female': 'Female',
    'mostly_female': 'Female',
    'andy': 'Unknown',
    'unknown': 'Unknown'
})
```

```python
df['Gender'] = df['Gender'].fillna(df['Gender_predicted'])
df
```

```python
df['Email'] = df['Email'].astype(str).str.lower().str.strip()  #str.lower()-converts all letters to lowercase,str.strip()-whitespce remove  
df['Email'] = df['Email'].str.replace(r'@+', '@', regex=True) #matches one or more '@' characters.replace them with a single @
df['Email'] = df['Email'].str.replace(r'@+', '@', regex=True) #second time check the @ are one are more replace the single  @
valid_email_pattern = r'^[a-z0-9._%+-]+@[a-z0-9.-]+\.[a-z]{2,}$' #correct position of email for regex.then domain name(letters,digits,hyphen,dot),must end with a dot and 2 or more lowercase letters.like,.com,.net,.org.
df = df[df['Email'].str.match(valid_email_pattern, na=False)]  # email pattern are match follow the check and remove invalid missing values

df.drop_duplicates(subset=['Email'], inplace=True) #more then duplictates are drop
df
```

```python
df['Email'] = df['Email'].str.replace(r'\.(biz|info|net)$', '.com', regex=True)
df
```

```python
df['PhoneNumber'] = df['PhoneNumber'].str.replace(r'\D', '', regex=True) #remove non-digit characters(spaces,symbols) from the 'phonenumber'colum,keeping only digits
df['PhoneNumber'] = df['PhoneNumber'].str.replace(r'^1', '', regex=True) #removes the leading '1' digit from phone numbers,if present.
df = df[df['PhoneNumber'].str.len() == 10] #keeps only those rows where the phone number length is exactly 10 digits
df
```

```python
df['TotalOrders']=pd.to_numeric(df['TotalOrders'],errors='coerce')
df
```

```python
df.isnull().sum()
```

```python
df["TotalOrders"]=df["TotalOrders"].astype(int)
```

```python
df
```

```python
df['TotalSpend']=df['TotalSpend'].astype(int)
df
```

```python
df['State']=df['State'].astype(str).str.replace(r'[^a-zA-Z]', '', regex=True).str.strip().str.title()
df       #r'[^a-zA-Z]', '', regex=True)---->remove all characters except english alphabets(both uppercase and lowercase)using regex.,str,strip()-->remove whitespace,str.title--->first letter ah uppercase word(change)and the rest to lowercase
```

```python
import matplotlib.pyplot as plt
```

```python
#barchart----> Gender
df['Gender'].value_counts().plot(kind='bar',color=['skyblue','salmon'])
plt.title('Gender Distribution')
plt.show()
```

```python
#histogram----> Age
plt.hist(df['Age'].dropna(),bins=10,color='teal')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
#boxplot---> Totalspend by customersegment
import seaborn as sns
sns.boxplot(x='CustomerSegment',y='TotalSpend',data=df)
plt.title('Spend Distribution by segment')
plt.show()
```

```python
#scatterplot--->Age,totalspend
plt.scatter(df['Age'],df['TotalSpend'],alpha=0.6)
plt.xlabel('Age')
plt.ylabel('TotalSpend')
plt.title('Age VS TotalSpend')
plt.show()
```

```python
#pie chart--->preferredpaymentmethod
df['PreferredPaymentMethod'].value_counts().plot(kind='pie',autopct='%1.1f%%')
plt.title('Payment method share')
plt.ylabel(' ')
plt.show()
```

```python
#Bar cahrt--->loyaltymember
df['LoyaltyMember'].value_counts().plot(kind='bar',color='orange')
plt.title('Loyalty membership status')
plt.show()
```

```python
#Time series line chart--->Registration date & Last purcahse date
df['RegistrationDate']=pd.to_datetime(df['RegistrationDate'],errors='coerce')
df.groupby(df['RegistrationDate'].dt.to_period('M'))['TotalSpend'].sum().plot(kind='area',color='pink')
plt.title('Monthly orders Trend')
plt.show()
```

```python
#Heatmap--->Age,Totalspend,totalorders,feedback score
corr=df[['Age','TotalSpend','TotalOrders','FeedbackScore']].corr()
sns.heatmap(corr,annot=True,cmap='coolwarm')
plt.title('correlation heatmap')
plt.show()
```

```python
#countplot--->customersegment
sns.countplot(x='CustomerSegment',data=df,order=['Bronze','Silver','Gold','Platinum'],palette='viridis')
plt.title('customer by segment')
plt.show()
```

```python
#Geo distribution--->country
Country_count=df['Country'].value_counts().head(10)
Country_count.plot(kind='bar',color='lightgreen')
plt.title('Top 10 countries by customer count')
plt.show()
```
